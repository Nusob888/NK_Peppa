```python
## Define function to assess gene variation

def vars(a, axis=None):
    """ Variance of sparse matrix a
    var = mean(a**2) - mean(a)**2
    """
    a_squared = a.copy()
    a_squared.data **= 2
    return a_squared.mean(axis) - np.square(a.mean(axis))

##trim the genes to remove non coding
#with open("/well/immune-rep/users/vbw431/Projects/Fairfax/Mel_all/data/coding_genes_noIg.txt") as f:
#    coding = f.read().splitlines() 

```

```python
##clean and assign hvg + subset

for name, value in adata_list.items():
    ##now trim genes for those present in more than 10 cells
    n_genes_pre = adata_list[name].shape[1]
    sc.pp.filter_genes(adata_list[name], min_cells=10)
    n_genes_post = adata_list[name].shape[1]
    print("Number of genes removed: " + str(n_genes_pre - n_genes_post))
    print("Number of genes pre-filtering: " + str(n_genes_pre))
    print("Number of genes post filtering: " + str(n_genes_post))
    
    ##Calculate hvg
    ## Visualise gene variation and dispersion
    means = np.mean(adata_list[name].X, axis=0)
    variances = vars(adata_list[name].X, axis=0)
    dispersions = variances / means 
    min_mean = 0.05

    fig, ax = plt.subplots(figsize=(6, 6))
    plt.scatter(
        np.log1p(means).tolist()[0], np.log(dispersions).tolist()[0], s=2
    )
    plt.vlines(x=np.log1p(min_mean),ymin=-2,ymax=8,color='red')
    plt.xlabel("log1p(mean)")
    plt.ylabel("log(dispersion)")
    plt.title(name+" DISPERSION VERSUS MEAN")
    plt.show()
    fig.savefig(plot_dir + name + "_gene_variance.png", bbox_inches="tight", dpi=300)
    
    ##Run HVG with subset
    sc.pp.highly_variable_genes(
        adata_list[name],
        n_top_genes=3000,
        batch_key="scvi_batch",
        flavor = "seurat_v3",
        subset=False)      
    
    
    fig, ax = plt.subplots(figsize=(6, 6))
    boolean_to_color = {
        True: "crimson",
        False: "steelblue",
    }  # make a dictionary that translates the boolean to colors
    hvg_colors = adata_list[name].var.highly_variable.map(boolean_to_color)  # 'convert' the boolean
    # now plot
    plt.scatter(
        np.log1p(means).tolist()[0], np.log(dispersions).tolist()[0], s=1, c=hvg_colors
    )
    plt.xlabel("log1p(mean)")
    plt.ylabel("log(dispersion)")
    plt.title(name+" DISPERSION VERSUS MEAN")
    plt.show()
    fig.savefig(plot_dir + name + "_HVGselect_gene_variance.png", bbox_inches="tight", dpi=300)
    
    
    ##Run HVG with subset
    sc.pp.highly_variable_genes(
        adata_list[name],
        n_top_genes=3000,
        batch_key="scvi_batch",
        flavor = "seurat_v3",
        subset=True)  
    
    ig_genes = []
    for gene in adata_list[name].var_names:
        m = re.search('.*(^IGH|^IGK|^IGV).*', gene)
        if m:
            ig_genes.append(gene)  
            
    #Save HVG
    if name == "Bcells":
        
        tokeep = adata_list[name].var_names[~adata_list[name].var_names.isin(ig_genes)]
        adata_list[name] = adata_list[name][:, tokeep]
        
        hvg = adata_list[name].var_names
        with open(r'/well/immune-rep/users/vbw431/Projects/Peppa/data/hvg_'+name+'_peppa.txt', 'w') as fp:
            for item in hvg:
                fp.write("%s\n" % item)
        
        print('Done '+name+' genes')
                
    else:
        hvg = adata_list[name].var_names
        with open(r'/well/immune-rep/users/vbw431/Projects/Peppa/data/hvg_'+name+'_peppa.txt', 'w') as fp:
            for item in hvg:
                fp.write("%s\n" % item)
        
        print('Done '+name+' genes')
    

```

## Loop train SCVI with scArches params

```python
for name, value in adata_list.items():
    ## setup model parameters 
    adata_list[name].layers["counts"] = adata_list[name].X.copy()
    outfile = "/well/immune-rep/users/vbw431/Projects/Peppa/scvi/"+name+"_model"
    
    ## train model
    model_wrappers.train_scVI(
        train_adata = adata_list[name],
        outfile = outfile,
        vae_epochs = 500,
        vae_batch_size = 256,
        n_latent = 30,
        batch = "scvi_batch"
    )
```

```python
##load vaes
TNK_vae = scvi.model.SCVI.load("/well/immune-rep/users/vbw431/Projects/Peppa/scvi/TNK_model")
Bcells_vae = scvi.model.SCVI.load("/well/immune-rep/users/vbw431/Projects/Peppa/scvi/Bcells_model")
MyePlt_vae = scvi.model.SCVI.load("/well/immune-rep/users/vbw431/Projects/Peppa/scvi/Myeloid_Plt_model")

vae_list = {"TNK": TNK_vae, "Bcells": Bcells_vae, "Myeloid_Plt": MyePlt_vae}
```

```python
for name, value in vae_list.items():
    #plot training
    train_test_results = vae_list[name].history['elbo_train']
    train_test_results['elbo_validation'] = vae_list[name].history['elbo_validation']
    train_test_results.plot(logy=True)
    plt.title(name+" model training")
    plt.show()
```

```python
for name, value in adata_list.items():
    print('generating latent space for '+name)
    adata_list[name].obsm["X_scvi"] = vae_list[name].get_latent_representation()
    
    print('generating neighbours for '+name)
    sc.pp.neighbors(adata_list[name], use_rep="X_scvi", metric="cosine",n_neighbors=30)
    
    print('generating umap for '+name)
    sc.tl.umap(adata_list[name])
    
    print('normalising data for '+name)
    sc.pp.normalize_total(adata_list[name], target_sum=1e4)
    sc.pp.log1p(adata_list[name])
    
    print('plotting umap embedding for '+name)
    with plt.rc_context():
        sc.pl.umap(
            adata_list[name],
            color=["Ref_lab", "batch","percent.mt", "leiden_4", "majority_voting"],
            frameon=False,
            #ncols=1,
        )
        plt.savefig(plot_dir+name+"model_umap.png", bbox_inches="tight", dpi=300)
```

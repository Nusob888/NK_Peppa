```python
for name in ["NK", "CD8", "CD4", "Bcells", "Myeloid_Plt"]:
    conditions = [
        (adata_list[name].obs["orig.ident"] == "SeuratProject"), ##absolute quality thresholds
        (adata_list[name].obs["orig.ident"] != "SeuratProject"), 
    ]
    #create a list of the values we want to assign for each condition

    values = [adata_list[name].obs["sample_ID"], adata_list[name].obs["lane"]]
    
    adata_list[name].obs["sequencing_batch"] = np.select(conditions, values)

    print(adata_list[name].obs["sequencing_batch"].value_counts())
```

```python
## Define function to assess gene variation

def vars(a, axis=None):
    """ Variance of sparse matrix a
    var = mean(a**2) - mean(a)**2
    """
    a_squared = a.copy()
    a_squared.data **= 2
    return a_squared.mean(axis) - np.square(a.mean(axis))

##trim the genes to remove non coding
with open("/well/immune-rep/users/vbw431/Projects/Fairfax/Mel_all/data/coding_genes_noIg.txt") as f:
    coding = f.read().splitlines() 

```

```python
##clean and assign hvg + subset

for name in ["NK", "CD8","CD4", "Bcells", "Myeloid_Plt"]:
    #adata_list[name] = adata_list[name][adata_list[name].obs["Ref_lab"]!= "COMBAT"].copy()
    #if name == "CD8NK":
        #adata_list[name] = adata_list[name][adata_list[name].obs["scvi_batch"].isin(["HBV_X2_CITE", "P4_0"])].copy()
    #else:
        #adata_list[name] = adata_list[name][adata_list[name].obs["scvi_batch"].isin(["HBV_X2_CITE"])].copy()
        
    ##now trim genes for those present in more than 5 cells
    #n_genes_pre = adata_list[name].shape[1]
    #sc.pp.filter_genes(adata_list[name], min_cells=5)
    #n_genes_post = adata_list[name].shape[1]
    #print("Number of genes removed: " + str(n_genes_pre - n_genes_post))
    #print("Number of genes pre-filtering: " + str(n_genes_pre))
    #print("Number of genes post filtering: " + str(n_genes_post))
    
    ##Calculate hvg
    ## Visualise gene variation and dispersion
    means = np.mean(adata_list[name].X, axis=0)
    variances = vars(adata_list[name].X, axis=0)
    dispersions = variances / means 
    min_mean = 0.05

    fig, ax = plt.subplots(figsize=(6, 6))
    plt.scatter(
        np.log1p(means).tolist()[0], np.log(dispersions).tolist()[0], s=2
    )
    plt.vlines(x=np.log1p(min_mean),ymin=-2,ymax=8,color='red')
    plt.xlabel("log1p(mean)")
    plt.ylabel("log(dispersion)")
    plt.title(name+" DISPERSION VERSUS MEAN")
    plt.show()
    fig.savefig(plot_dir + name + "_gene_variance.png", bbox_inches="tight", dpi=300)
    
    ##Run HVG with subset
    sc.pp.highly_variable_genes(
        adata_list[name],
        n_top_genes=3000,
        batch_key="sequencing_batch",
        flavor = "seurat_v3",
        subset=False)      
    
    
    fig, ax = plt.subplots(figsize=(6, 6))
    boolean_to_color = {
        True: "crimson",
        False: "steelblue",
    }  # make a dictionary that translates the boolean to colors
    hvg_colors = adata_list[name].var.highly_variable.map(boolean_to_color)  # 'convert' the boolean
    # now plot
    plt.scatter(
        np.log1p(means).tolist()[0], np.log(dispersions).tolist()[0], s=1, c=hvg_colors
    )
    plt.xlabel("log1p(mean)")
    plt.ylabel("log(dispersion)")
    plt.title(name+" DISPERSION VERSUS MEAN")
    plt.show()
    fig.savefig(plot_dir + name + "_HVGselect_gene_variance.png", bbox_inches="tight", dpi=300)
    
    
    ##Run HVG with subset
    sc.pp.highly_variable_genes(
        adata_list[name],
        n_top_genes=3000,
        batch_key="sequencing_batch",
        flavor = "seurat_v3",
        subset=True)  
    
    ig_genes = []
    for gene in adata_list[name].var_names:
        m = re.search('.*(^IGHV|^IGK|^IGL|^IGHJ).*', gene)
        if m:
            ig_genes.append(gene)  
            
    #Save HVG
    if name == "Bcells":
        
        tokeep = adata_list[name].var_names[~adata_list[name].var_names.isin(ig_genes)]
        adata_list[name] = adata_list[name][:, tokeep]
        
        hvg = adata_list[name].var_names
        #hvg = adata_list[name].var_names[adata_list[name].var_names.isin(coding)]
        with open(r'/well/immune-rep/users/vbw431/Projects/Peppa/data/hvg_'+name+'_peppa.txt', 'w') as fp:
            for item in hvg:
                fp.write("%s\n" % item)
        
        print('Done '+name+' genes')
                
    else:
        hvg = adata_list[name].var_names
        #hvg = adata_list[name].var_names[adata_list[name].var_names.isin(coding)]
        with open(r'/well/immune-rep/users/vbw431/Projects/Peppa/data/hvg_'+name+'_peppa.txt', 'w') as fp:
            for item in hvg:
                fp.write("%s\n" % item)
        
        print('Done '+name+' genes')
    

```
